```python
import pandas as pd
import numpy as np

param_headers=[]
param_dim=12
stage_dim=11
dev_dim=3

for i in range(param_dim):
    param_headers.append("pp_"+str(i+1))

param_headers
```

```python
#Reading Process Parameters
param_df = pd.read_csv("cross_member_samples_test1.csv",header=None,names=param_headers)
print("Data Shape : ", param_df.shape)
```

```python
moving_window=20

extracted_temporal_data=param_df.iloc[-20:,:]
extracted_temporal_data
```

```python
#Out of control Check for Process Parameters

#Dummy control limits, to be taken as input 
pp_lcl=[-1]*param_dim
pp_ucl=[1]*param_dim

ooc_violations=[]
for i,col in enumerate(extracted_temporal_data.columns):
    #Extracted Check
    ooc_violations.append(extracted_temporal_data[(extracted_temporal_data[col] > pp_ucl[i]) | (extracted_temporal_data[col] < pp_lcl[i])][[col]])

ooc_violations[i]
```

```python
# Checking for mean shift in process
# One sample T-Test
def one_ttest(data,mean_0=0,sig=0.05):
    from scipy.stats import ttest_1samp
    tset, pval = ttest_1samp(data, mean_0)
    
    test_flag=0
    #print(pval)
    if(pval < sig):
        test_flag=1
    
    return test_flag,pval

mean_vector=[0]*param_dim
process_mean=extracted_temporal_data.mean()
process_std=extracted_temporal_data.std()
#Extracted Process Mean and Standard Deviation
#print(process_mean,process_std)

mean_shift_violations=[]
for i,col in enumerate(extracted_temporal_data.columns):
    #Extracted Check
    mean_shift_violations.append(one_ttest(extracted_temporal_data[col].values,mean_vector[i]))

mean_shift_violations
```

```python
# Checking for drift, stationarity
def adf_test(data,sig=0.05):    
    from statsmodels.tsa.stattools import adfuller
    stat, pval, lags, obs, crit, t = adfuller(data)   
    test_flag=0
    
    #print(pval)
    if(pval < sig):
        test_flag=1
    
    return test_flag,pval

sta_violations=[]
for i,col in enumerate(extracted_temporal_data.columns):
    #Extracted Check
    sta_violations.append(adf_test(extracted_temporal_data[col].values))

sta_violations
```

```python
#Check for heteroskedascity
def lmarch_test(data,sig=0.05):
    
    from statsmodels.stats.diagnostic import het_arch
    stat,pval,stat_f,pval_f=het_arch(data)
    test_flag=0
    
    #print(pval)
    if(pval < sig):
        test_flag=1
    
    return test_flag,pval

het_violations=[]
for i,col in enumerate(extracted_temporal_data.columns):
    #Extracted Check
    het_violations.append(lmarch_test((extracted_temporal_data[col]-extracted_temporal_data[col].mean()).values))

het_violations
```

```python

def get_reward_signal(reward, loss_thres=-0.9,los_neg_thres=-3):
    
    #less_thres can be annleaded with time for continous improvement
    reward_signal=0.0
    
    #Scaling Rewards Between 0 to 1
    
    if(reward>loss_thres):
        reward_signal=1.0
    elif(reward<los_neg_thres):
        reward_signal=0.0
    else:
        reward_signal=(((reward-los_neg_thres)/(loss_thres-los_neg_thres))*1)
    
    return reward_signal
    
```

```python
from sklearn.datasets import make_classification
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import shuffle
import numpy as np

X, y1 = make_classification(n_samples=5, n_features=5, n_informative=2, n_classes=2, random_state=1)
y2 = shuffle(y1, random_state=1)
Y = np.vstack((y1, y2)).T

forest = RandomForestClassifier(n_estimators=10, random_state=1)
multi_target_forest = MultiOutputClassifier(forest, n_jobs=-1)
multi_target_forest.fit(X, Y).predict(X)

print(multi_target_forest.predict_proba(X))
```
